```python
# Instalamos el programa
!pip install gtts
from gtts import gTTS
from IPython.display import Audio
```

PARÁMETROS QUE SE UTILIZAN

**language** : es el idioma en el que se quiere convertir el texto en habla

**dialecto_diferenciado** : es el dialecto específico que se utilizará: España= es, USA = us, Mexico = com.mx

**text** : es el texto a sintetizar

**slow_par** : puede ser False (por defecto) o True (pronunciación más lenta)





```python
# Introducir abajo los parámetros
language = 'es'
dialecto_diferenciado = 'es'
slow_par = False
```

```python

# Introduzca el texto abajo sustituyendo las palabras que están entre comillas

text= "En un lugar de la Mancha, de cuyo nombre no quiero acordarme, no ha mucho tiempo que vivía un hidalgo de los de lanza en astillero, adarga antigua, rocín flaco y galgo corredor. Una olla de algo más vaca que carnero, salpicón las más noches, duelos y quebrantos los sábados, lentejas los viernes, algún palomino de añadidura los domingos, consumían las tres partes de su hacienda"
```

```python
# Generación de la voz
speech = gTTS(text= text, lang= language, slow = slow_par, tld= dialecto_diferenciado)
# guardo la voz en un fichero
speech.save("Quijote_normal.mp3")
# reproduzco el fichero
Audio("Quijote_normal.mp3")

```

**Respuesta:**
Después de escuchar la grabación, vemos que hay ciertos factores que afectan a la prosodia.

En primer lugar, al no tener signos de puntuación, es el sistema el que tiene que decidir por sí mismo dónde realizar las pausas. El texto está compuesto únicamente por dos frases largas. La máquina ha realizado pausas en lugares donde ninguna persona las realizaría, separando elementos que forman parte de un mismo grupo entonativo. Por ejemplo:
* hidalgo / de los de lanza en astillero: está separando el núcleo del sintagma nominal de sus complementos.
* de algo mas vaca / que carnero: aquí separa los dos términos de la comparación
* algun / palomino de añadidura: en este caso ha separado el determinante del nombre que determina.

En ninguno de estos tres casos debería existir una pausa, sin embargo, son los únicos tres momentos en los que el sistema decide 'coger aire'.

Por otro lado, tampoco sabe cómo realizar las enumeraciones. Aunque en este caso no se debe hacer una pausa por cada coma que hubiera entre elemento y elemento de la enumeración, sí que deberían existir pequeñas modulaciones en la entonación que el sistema no es capaz de realizar.

Finalmente, aunque no tenga signos de puntuación, el sistema debería ser capaz de detectar los grupos entonativos. Por ejemplo, después de 'de cuyo nombre no quiero acordarme', debería realizar una pequeña subida del tono para indicar que la frase aún no ha terminado. Sin embargo, no realiza las modulaciones necesarias.

Todo esto dificulta la comprensión del mensaje por parte del receptor.

B.   Buscar ejemplos en los que la pronunciación de números no sería la más utilizada por la gente en general. En ese caso, plantee una posible solución




```python
text = """
Estamos en el año 2024.
Mi DNI es 32894306.
Su DNI es 32894306J.
Nos vemos mañana a las 21:45.
El número de teléfono de la empresa es 639171358.
Cuesta 300€.
Tengo 300 monedas."""

speech = gTTS(text= text, lang= language, slow = slow_par, tld = dialecto_diferenciado)
speech.save("B.mp3")
Audio("B.mp3")
```

**Respuesta:**
Los sistemas de conversión texto habla tienen que tener en cuenta las convenciones sociales que existen alrededor de la pronunciación de multitud de expresiones, véase abreviaciones, siglas, etc. Los números son una de ellas, la manera de decirlos dependerá del contexto y de la región. A continuación muestro algunos ejemplos donde la pronunciación del sistema puede no ser la más utilizada:

1. **Años:** en español, los años se pronuncian de la misma manera que el número completos (aquí 'dos mil veinticuatro'). Sin emabargo en otros idiomas, como el inglés, se pronuncian por parejas ('veinte veinticuatro'). El sistema debe ser consciente de estas variaciones y adecuarse a cada idioma.

2. **DNI:** este caso resulta curioso ya que si escribimos el DNI sin la letra, lo pronuncia como un número entero ('treinta y dos millones...'). Sin embargo, cuando añadimos la letra final, lo pronuncia dígito a dígito. En España es común escuchar ambas versiones del DNI, por lo que considero que las dos pronunciaciones serían correctas.

3. **Horas:** en este caso le hemos dado la hora en formato digital. El sistema, muy adecuadamente sabe que a partir de las 12 se considera horario de tarde, por lo que ha pronunciado las 'nueve y cuarenta y cinco de la noche'. No obstante, también sería correcta la versión más literal 'las veintiuno cuarenta y cinco' o la versión más completa posible: 'las diez menos cuarto de la noche'. Quizás el sistema podría tener registradas varias versiones de entre las cuales el ususario pudiera elegir a su gusto.

4. **Números de teléfono:** en este caso el sistema se equivoca en gran manera, ya que pronuncia el número de teléfono como si fuera un único número. Hay muchas maneras en la que la gente dice el número de teléfono, algunos dígito a dígito, otros mezclando dígitos con grupos de números o haciendo solo agrupaciones de números. Para que pronuncie dígito a dígito podemos escribir un espacio entre número y número; para que pronuncie distintos grupos de números los separamos entre comas. De todas maneras, la mejor manera de que el sistema corrija esto es explicitando en el preprocesamiento de qué manera tendría que pronunciarlos.

5. **Cantidades de dinero:** parece que el sistema es bastante preciso coordinando el género de los números con el sustantivo cuando se trata de divisas (ej. 'trescient**o**s euros'). Sin embargo, cuando hablamos de un sustantivo común, como puede ser 'monedas', no es capaz de cambiar el género de 'trescientos' a femenino. En este caso de nuevo, habría que añadir un paso en el preprocesamiento para que se adaptara en estos casos.

**Respuesta:**
Las diferencias más claras a nivel fonético entre estos tres dialectos son:

1. **Seseo**: en el dialecto de españa la "z" la "c" y la "s" son fonemas distintos, aunque haya regiones en las que exista el seseo o el ceceo. Sin embargo, en el español mejicano y el de Estados Unidos, no tienen distintción entre "z", "c" y "s", todos los pronuncian como "s". Por tanto, en los audios mejicano y estadounidense, escuchamos "selebrar" y en el español "celebrar".

2. **Aspiraciones**: en los dialectos mejicano y de Estados Unidos, se pronuncia "hardín", con una aspiración inicial donde tendría que estar el alófono de la "j". Sin embargo, en español, es un sonido mucho más fuerte y marcado.

3. **Pronunciación de las vocales:** en español de España, las vocales tienden a ser más cerradas y pueden sonar más "redondeadas" en comparación con el español de México. Por ejemplo, la "e" en "celebrar" y "abuela" suena más cerrada en español de España que en español de México.

4. **Pronunciación de la "r":** en español de España, la "r" tiende a ser más vibrante, mientras que en el español de México puede ser más suave o incluso aspirada. Por ejemplo, la "r" en "celebrar" suena más vibrante en español de España y más suave en español de México.

5. **Prosodia**: en español de España, la entonación puede ser más marcada, mientras que en español de México puede ser más plana.

Es importante tener en cuenta que estas diferencias pueden variar dependiendo de la región específica dentro de cada país, ya que tanto España como México tienen una gran diversidad lingüística dentro de sus territorios.
